# add_employee_link

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# Read the Excel file
excel_data = pd.read_excel('sample_data.xlsx')

# Initialize the web driver
driver = webdriver.Chrome()
driver.maximize_window()

# Navigate to your web application
driver.get('http://127.0.0.1:5000/')

# Loop through the rows in the Excel file
for index, row in excel_data.iterrows():
    #time.sleep(1)
    # Click on the "Add Employee" link
    add_employee_link = driver.find_element(By.XPATH, '//ul/li/a[@href="/add_employee"]')
    add_employee_link.click()
    time.sleep(1)  # Wait for the page to load
    
    # Locate the form fields using Selenium
    id_field = driver.find_element(By.XPATH, '//*[@id="id"]')
    name_field = driver.find_element(By.XPATH, '//*[@id="name"]')
    position_field = driver.find_element(By.XPATH, '//*[@id="position"]')
    salary_field = driver.find_element(By.XPATH, '//*[@id="salary"]')
    
    # Clear the form fields (optional)
    id_field.clear()
    name_field.clear()
    position_field.clear()
    salary_field.clear()
    
    # Fill in the form fields with data from the Excel file
    id_field.send_keys(str(row['ID']))
    name_field.send_keys(row['Name'])
    position_field.send_keys(row['Position'])
    salary_field.send_keys(str(row['Salary']))
    
    # Submit the form
    submit_button = driver.find_element(By.XPATH, '/html/body/form/button')
    submit_button.click()
    time.sleep(1)  # Wait for the form to be submitted and the page to load
    
    submit_button = driver.find_element(By.XPATH, '/html/body/a')
    submit_button.click()
    time.sleep(1)

# Close the browser
driver.quit()
```

# REMOVE EMPLOYEE

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Read the Excel file
excel_data = pd.read_excel('sample_data.xlsx')

# Initialize the web driver
driver = webdriver.Chrome()
driver.maximize_window()

# Navigate to your web application
driver.get('http://127.0.0.1:5000/')

# Loop through the rows in the Excel file to remove each employee
for index, row in excel_data.iterrows():
    # Wait for and click on the "Remove Employee" link
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '/html/body/ul/li[2]/a'))
    ).click()
    
    # Wait for the form field to be present
    id_field = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="id"]'))
    )
    
    # Clear the form field (optional)
    id_field.clear()
    
    # Fill in the form field with the employee ID from the Excel file
    id_field.send_keys(str(row['ID']))
    
    # Wait for and click the "Remove" button
    remove_button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '/html/body/form/button'))
    )
    remove_button.click()
    
    # Wait for and click the "Home" button
    home_button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '/html/body/a'))
    )
    home_button.click()
    time.sleep(1)  # Adjust the sleep time if necessary

# Close the browser
driver.quit()
```
